## フォルダ内のマルチページtiff複数を，複数pngに変換して書き出し

```python
def resizeImg(mm):
    SHORT_EDGE = 384
    h,w = mm.shape[:2]
    if w>h: # 横長のとき，高さを384にして，横を同アスペクト比で調整．
        mm = cv2.resize(mm, dsize=(round(SHORT_EDGE*w/h), SHORT_EDGE))
    else:
        mm = cv2.resize(mm, dsize=(SHORT_EDGE, round(SHORT_EDGE*h/w)))
    return mm

def writePng(imgFileName:str, writePath:str) -> None:
    myImgMat = []
    IMGNUM_MAX = 20
    writeDir = writePath
    filepath = imgFileName
    _, myImgMat = cv2.imreadmulti(filename = filepath, start=1, count=IMGNUM_MAX, flags = cv2.IMREAD_ANYCOLOR)

    MatCnt = 0
    for mm in myImgMat:
        MatCnt += 1
        basename_without_ext = os.path.splitext(os.path.basename(filepath))[0]
        writeName = basename_without_ext + "_fig" + str(MatCnt) + ".png"
        img_file = os.path.join(writeDir, writeName)
        #print(img_file)
        mm = resizeImg(mm)
        cv2.imwrite(img_file, mm)
```
